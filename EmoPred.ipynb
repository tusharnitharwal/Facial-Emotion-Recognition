```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./feruse.csv')
```

```python
print(df.head)
```

```python
df['emotion'].head
```

```python
X_test, X_train, y_test, y_train = [], [] ,[], []
```

```python
for index, row in df.iterrows():
    val = row['pixels'].split(" ")    
    if 'Training' in row['Usage']:
        X_train.append(np.array(val, dtype = np.float32))
        y_train.append(row['emotion'])
    elif 'PublicTest' in row['Usage']:
        X_test.append(np.array(val, dtype = np.float32))
        y_test.append(row['emotion'])
```

```python
X_test
```

```python
X_train = np.array(X_train, dtype = 'float32')
y_train = np.array(y_train, dtype = 'float32')
X_test = np.array(X_test, 'float32')
y_test = np.array(y_test, dtype = 'float32')
```

```python
X_test.shape
```

```python
X_train -= np.mean(X_train, axis=0)
X_train /= np.std(X_train, axis=0)

X_test -= np.mean(X_test, axis=0)
X_test /= np.std(X_test, axis=0)

# X_train /= 255
# X_test/= 255
```

```python
X_train = X_train.reshape(X_train.shape[0], 48, 48, 1)
```

```python
X_train.shape
```

```python
X_test = X_test.reshape(X_test.shape[0], 48, 48, 1)
```

```python
import tensorflow as tf
from tensorflow.keras import utils
```

```python
y_train[0:10]
```

```python
y_train = utils.to_categorical(y_train, num_classes = 7)
y_test =  utils.to_categorical(y_test, num_classes = 7)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array

datagen = ImageDataGenerator(rotation_range = 40, zoom_range = 0.5, width_shift_range = 0.2, height_shift_range = 0.2, 
                             shear_range = 0.2, horizontal_flip = True, vertical_flip = True, fill_mode = "nearest")

datagen.fit(X_train)
```

```python
model.compile( optimizer = tf.keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-7), loss = tf.keras.losses.categorical_crossentropy, metrics = ['accuracy'])
```

```python
model.fit_generator(datagen.flow(X_train, y_train, batch_size =32),steps_per_epoch = len(X_train)/32, epochs = 40, verbose = 1,  validation_data=(X_test, y_test ), shuffle = True)
```
